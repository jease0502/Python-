```python
while(True): 
    url = "https://www.cec.gov.tw/pc/zh_TW/P1/n00000000000000000.html"
    resp = requests.get(url,verify = False)
    resp.encoding = 'utf-8'
    html = BeautifulSoup(resp.text, "lxml")

    content = html.find(name='table', attrs={'cellpadding':'6'})
    i=0
    number = [0,0,0]
    for data in content.find_all(name='tr',attrs = {'class':'trT'}):
        number[i]=data.find(name='td',attrs = {'class':'tdAlignRight'}).text
        number[i] = number[i].replace(',', '')
        i+=1
    print(number)
    picc = content.find(name='td',attrs = {'align':'right'})
    print(picc.text)
    plt.figure(figsize=(6,9))
    size = number

    x = ['宋楚瑜', '韓國瑜', '蔡英文']
    pie_color = ["orange","blue", "green"]

    plt.pie(number, colors = pie_color, labels = x , autopct = "%2.2f%%")
    plt.title("2020 總統大選")
    plt.pause(0.05)
    time.sleep(90)
    plt.clf()
```
